# Data Storytelling
With the data I've just collected and cleaned, I want to be able to solve a huge problem that a lot of NFL teams face. They need a quarterback and need to make the best possible choice. Is there a surefire way to limit the probability of making a poor choice. Are there aspects of their careers that we can point to in order to determine whether or not they would make a good choice? And is there a way to determine whether a QB may become a "late-round steal" (A player that is drafted late that will go on to become a great player)?

I will now perform some exploratory analysis on the data to explore the dataset further.

```python
#Import appropriate packages
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
```

```python
#Import data
ncaa_stats = pd.read_csv('data/pfr_ncaa_draft_data_CLEAN.csv')
ncaa_stats
```

```python
#Explore numbers for the dataset
ncaa_stats.describe()
```

### Observing Distributions

```python
sns.set()
sns.distplot(ncaa_stats.career_av)
plt.xlim(-25, 200)
plt.title('Distribution of Career Approximate Value')
plt.show()
```

Of all of the quarterbacks selected in every NFL draft from 1970 through 2012, very few of them have high career Approximate Values (AV). This distribution centers around 0, which is very telling in itself. Perhaps even more telling - it has an extremely long tail.

After observing the distributions of all of the individual quantitative columns, I found that the following distributions interested me the most.

```python
sns.boxplot(x='pass_cmp_pct', data=ncaa_stats)
plt.title('Distribution of Career Passing Completion Percentage', fontsize=14)
plt.show()
```

Most players have a pass completion percentage of around 55%, which makes logical sense. What stands out in this box plot is the fact that there are some players who's completion percentage was at or around 100%! Let's take a look to see what that is about.

```python
ncaa_stats[ncaa_stats.pass_cmp_pct > 90]
```

It appears as though there were a handful of players with completion percentages at 100%, but it is because they only had less than 5 passing attempts. That makes much more sense.

```python
sns.boxplot(x='adj_pass_yds_per_att', data=ncaa_stats)
plt.title('Distribution of Career Adjusted Passing Yards/Attempt', fontsize=14)
plt.show()
```

This is another interesting distribution. The adjusted passing yards/attempt is an index that takes into account the number of touchdowns and interceptions (The formula is: Yds + 20 * TD - 45 * Int / Att). Most players see their number around 7 yards or below, but there are some outliers that have a much higher number than that.

```python
ncaa_stats[ncaa_stats.adj_pass_yds_per_att >= 10]
```

Again, we have some players here with low passing attemps, which can skew things a bit, but there are names like Sam Bradford, Cam Newton and Russell Wilson who have gone on to have great NFL careers.

```python
sns.boxplot(x='pass_td', data=ncaa_stats)
plt.title('Distribution of Career Passing Touchdowns', fontsize=14)
plt.show()
```

A large majority of the players are between 20 and 50 passing touchdowns. Again, I'm seeing some big outliers that I would like to explore further.

```python
ncaa_stats[ncaa_stats.pass_td >= 85]
```

As expected, in this list, we are able to see some fairly big names like Drew Brees, Peyton Manning and Phillip Rivers among others.

```python
sns.boxplot(x='pass_int', data=ncaa_stats)
plt.title('Distribution of Career Passing Interceptions', fontsize=14)
plt.show()
```

I am particularly intersted in the relationship between career AV and career interceptions. One would think that the higher the interceptions, the lower the career AV.

```python
sns.boxplot(x='rush_td', data=ncaa_stats)
plt.title('Distribution of Career Rushing Touchdowns', fontsize=14)
plt.show()
```

Similarly to passing touchdowns, I will take a look at the career rushing touchdown outliers a bit further.

```python
ncaa_stats[ncaa_stats.rush_td >= 30]
```

Interstingly, some of the names here are recognizable, but none of them had particularly notable careers in the NFL statistically. It makes me wonder if the number of rushing touchdowns in college is as valuable as the number of passing touchdowns (As far as translating that success in college to the NFL).

Specifically, Tim Tebow and Vince Young appear in this data frame. Both of them had incredible college careers, but were not considered very good quarterbacks in the NFL.

### Observing Scatterplots

I will now use Seaborn's pairplot tool to see how all of the quanitative variables relate to career AV.

```python
#Scatter plots for college passing stats vs. career AV
sns.set()
sns.pairplot(ncaa_stats, vars=['career_av', 'pass_cmp', 'pass_att','pass_cmp_pct', 'pass_yds', 'pass_yds_per_att', 
                               'adj_pass_yds_per_att', 'pass_td', 'pass_int'])              
plt.show()
```

Takeaways:

- There does not seem to be any discernable relationship between career AV and pass completion percentage, pass yards/attempt and adjusted pass yards/attempt
- The rest of the passing statistics all seem to have a very slightly positive relationship
- What we are gathering here, based just off of looking at these plots is that the relationship between career AV and these various passing statistics from the players' college careers is not very strong

```python
#Scatter plots for college rushing stats vs. career AV
sns.set()
sns.pairplot(ncaa_stats, vars=['career_av', 'rush_att', 'rush_yds', 'rush_yds_per_att', 'rush_td'])              
plt.show()
```

Between the rushing statistics and career AV, the relationship seems to be less positive than the passing statistics. However, this isn't entirely surprising considering the NFL is (And has historically been) mostly a passing league where rushing quarterbacks are not usually as successful. 

# Statistics

I will now take a deeper look at the relationship between career AV and the various college statistics we have available. I will do that by running a multivariate linear regression and checking the p-value for each variable to determine which ones I need and which ones I can ignore.

We will use a significance level of .05.

```python
# Designate response variable
y = ncaa_stats.career_av

# Designate explanatory variables
X = ncaa_stats[['pass_cmp', 'pass_att', 'pass_cmp_pct', 'pass_yds', 'pass_yds_per_att', 'adj_pass_yds_per_att', 
               'pass_td', 'pass_int', 'rush_att', 'rush_yds', 'rush_yds_per_att', 'rush_td']]

# Add constant
X = sm.add_constant(X)

# Run the model
reg = sm.OLS(y, X).fit()

# Show regression results
reg.summary()
```

```python
# Run model again without some of the variables with higher p-values
X = ncaa_stats[['pass_cmp', 'pass_att', 'pass_cmp_pct', 'pass_yds', 
                'pass_int', 'rush_att', 'rush_yds_per_att', 'rush_td']]
X = sm.add_constant(X)
reg1 = sm.OLS(y, X).fit()
reg1.summary()
```

```python
# Repeat
X = ncaa_stats[['pass_yds', 'rush_att', 'rush_yds_per_att', 'rush_td']]
X = sm.add_constant(X)
reg2 = sm.OLS(y, X).fit()
reg2.summary()
```

```python
# Final adjustment
X = ncaa_stats[['pass_yds', 'rush_att', 'rush_td']]
X = sm.add_constant(X)
reg3 = sm.OLS(y, X).fit()
reg3.summary()
```

Interestingly, our model suggests that passing yards, rushing attemps and rushing touchdowns are the most statisitcally significant variables included in the regression. However, those variables only make up around 5.7% of the total variance in career AV, which means that there are other variables (That we do not have access to) that are explaining the bulk of the variance for career AV.
